```python
import pandas as pd

df = pd.read_csv("IDS_ALLCountries_Data.csv", encoding="latin1")
print(df)
```

```python
df = df.dropna(how="all")
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
missing_percent = (df.isna().sum() / len(df)) * 100
print(missing_percent)
```

```python
df = df.drop_duplicates()
df
```

```python

cols_to_drop = [
 "2025","2026","2027","2028","2029","2030","2031","2032"
]

df = df.drop(columns=cols_to_drop)
```

```python
df
```

```python
cols_to_drop = [
 "Counterpart-Area Code","Counterpart-Area Name",
]

df = df.drop(columns=cols_to_drop)
```

```python
df.columns = df.columns.str.strip()

# melt (convert columns → rows)
df = pd.melt(
    df,
    id_vars=[
        "Country Name",
        "Country Code",
        "Series Name",
        "Series Code"
    ],
    var_name="Year",
    value_name="Value"
)
df
```

```python
df.isna().sum()
```

```python
df["Value"] = df["Value"].fillna(0)
```

```python
df = df.dropna(subset=[
    "Country Code",
    "Series Name",
    "Series Code"
])
```

```python
import seaborn as sns
```

```python
print(df.describe())
```

```python
total_global_debt = df["Value"].sum()

print("Total Global Debt:", total_global_debt)
```

```python
top_countries = (
    df.groupby("Country Name")["Value"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_countries)

plt.figure(figsize=(12,6))

top_countries.plot(kind="bar")

plt.title("Top 10 Countries with Highest Debt")
plt.xlabel("Country")
plt.ylabel("Total Debt")
plt.xticks()
plt.show()
```

```python
lowest_countries = (
    df.groupby("Country Name")["Value"]
    .sum()
    .sort_values()
    .head(10)
)

print(lowest_countries)
```

```python
yearly_debt = (
     df.groupby("Year")["Value"]
    .sum()
)

print(yearly_debt)

plt.figure(figsize=(12,6))

plt.plot(yearly_debt.index, yearly_debt.values)

plt.title("Global Debt Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Debt")
plt.grid(True)
plt.show()
```

```python
indicator_debt = (
    df.groupby("Series Name")["Value"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(indicator_debt)
plt.figure(figsize=(12,8))

indicator_debt.plot(kind="barh")

plt.title("Top 10 Debt Indicators")
plt.xlabel("Debt Value")
plt.ylabel("Indicator")
plt.show()
```

```python
avg_country_debt = (
    df.groupby("Country Name")["Value"]
    .mean()
    .sort_values(ascending=False)
)

print(avg_country_debt.head(10))
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(x=df["Value"], y=df["Year"], data=df)

plt.title("Outlier Detection in Debt Values")
plt.show()
```

```python
pivot_table = df.pivot_table(
    values="Value",
    index="Country Name",
    columns="Year",
    aggfunc="sum"
)

plt.figure(figsize=(15,10))

sns.heatmap(pivot_table.head(20), cmap="viridis")

plt.title("Country-wise Debt Heatmap")
plt.show()
```

```python
numeric_df = df[["Value", "Year"]]

correlation = numeric_df.corr()

print(correlation)
```
