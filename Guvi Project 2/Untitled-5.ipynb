```python
import pandas as pd

df=pd.read_csv("IDS_SeriesMetaData.csv", encoding="latin1")
df
```

```python
df.isna().sum()
```

```python
cols_to_drop = [
 "License Type","Limitations and exceptions","General comments"
]

df = df.drop(columns=cols_to_drop)
```

```python
df.info()
```

```python
df = df.ffill()
```

```python
df.isna().sum()
```

```python
print(
    "Unique Series Codes:",
    df["Code"].nunique()
)
```

```python
print(
    "Duplicate Rows:",
    df.duplicated().sum()
)
```

```python
topic_freq = (
    df["Topic"]
    .value_counts()
)

print(topic_freq)
```

```python
import matplotlib.pyplot as plt
```

```python
topic_freq.plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("Topic Distribution")
plt.ylabel("Count")


plt.show()
```

```python
df["Long_Name_Length"] = (
    df["Long definition"]
    .astype(str)
    .apply(len)
)

print(
    df["Long_Name_Length"]
    .describe()
)
```

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["Long_Name_Length"],
    bins=30
)

plt.title("Long Definition Length Distribution")

plt.show()
```

```python
short_def = (
    df["Short definition"]
    .value_counts()
    .head(10)
)

print(short_def)
```

```python
source_freq = (
    df["Source"]
    .value_counts()
    .head(10)
)

print(source_freq)

source_freq.plot(
    kind="barh",
    figsize=(22,16)
)

plt.title("Top Metadata Sources")

plt.show()
```

```python
null_percent = (
    df.isna().sum() / len(df)
) * 100

print(
    null_percent.sort_values(
        ascending=False
    )
)
```

```python
cross = pd.crosstab(
    df["Topic"],
    df["Source"]
)

print(cross.head())
```

```python
topic_percent = (
    df["Topic"]
    .value_counts(normalize=True)
) * 100

print(topic_percent)
```

```python
topic_percent.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(8,8)
)

plt.title("Topic Percentage Distribution")

plt.ylabel("")

plt.show()
```
